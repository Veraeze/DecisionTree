```python
# import necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# data visualization settings

sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# load the data

from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
dataframe = pd.DataFrame(data.data, columns=data.feature_names)
dataframe['target'] = data.target
```

```python
# overview of data

print(dataframe.head())
print(dataframe.info())
print(dataframe.describe())
print(dataframe.shape)
```

```python
# plot count of tumor classes

sns.countplot(x='target', data=dataframe)
plt.title('Distribution of Tumor classes')
plt.xticks([0, 1], ['Benign', 'Malignant'])
plt.show()
```

```python
# scale features to avoid lbfgs solver failure

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(data.data)
```

```python
# seperate the dataset into training set and test set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, data.target, test_size=0.2, random_state=42)
```

```python
# fit the trainig set into the model

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
# make predictions and evaluate

y_pred = model.predict(X_test)
y_pred
```

```python
# check for model accuracy

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# plot confusion matrix 

from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_estimator(model, X_test, y_test)
plt.title("Confusion Matrix")
plt.show()
```

```python
# plot ROC curve

from sklearn.metrics import RocCurveDisplay

RocCurveDisplay.from_estimator(model, X_test, y_test)
plt.title("ROC Curve")
plt.show()
```

```python
# plot precision - recall curve

from sklearn.metrics import PrecisionRecallDisplay

PrecisionRecallDisplay.from_estimator(model, X_test, y_test)
plt.title("Precision-Recall Curve")
plt.show()
```

```python
# plot top 5 features that increase the probability of being malignant

feature_importance = pd.Series(model.coef_[0], index=data.feature_names)
feature_importance.nlargest(5).plot(kind='barh')
plt.title("Top 5 Important Features")
plt.xlabel("Coefficient Value")
plt.show()
```

```python
# plot top 5 features that reduce the probability of being spam

feature_importance = pd.Series(model.coef_[0], index=data.feature_names)
feature_importance.nsmallest(5).plot(kind='barh', color='green')
plt.title("Top 5 Important Features")
plt.xlabel("Coefficient Value")
plt.show()
```
